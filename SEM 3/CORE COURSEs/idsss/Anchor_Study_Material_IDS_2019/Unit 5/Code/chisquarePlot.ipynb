```python
%matplotlib inline
import numpy as np
from scipy.stats import chi2
from scipy.stats import chi
from matplotlib import pyplot as plt



# Define the distribution parameters to be plotted
k1_values = [1, 2]
k2_values = [3, 6, 12, 24]

mu = 0
x = np.linspace(0, 50, 1000)
color = ['red', 'black', 'green','blue','gray']

# plot the distributions


for k,c in zip(k1_values, color):
    dist = chi2(k, mu)
    plt.plot(x, dist.pdf(x), c=c,label=r'$k=%i$' % k)

plt.xlim(0, 10)
plt.ylim(0, 1)

plt.xlabel('Q')
plt.ylabel('p(Q|k)')
plt.title('$\chi^2\ \mathrm{Distribution}$')

plt.legend()
plt.show()



for k,c in zip(k2_values, color):
    dist = chi2(k, mu)
    plt.plot(x, dist.pdf(x), c=c,label=r'$k=%i$' % k)

plt.xlim(0, 50)
plt.ylim(0, 0.25)

plt.xlabel('Q')
plt.ylabel('p(Q|k)')
plt.title('$\chi^2\ \mathrm{Distribution}$')

plt.legend()
plt.show()
```

```python
%matplotlib inline
import numpy as np
from scipy.stats import chi2
from matplotlib import pyplot as plt

x = np.arange(-1,200)
#df = n - 1
for n in [2, 7, 13, 30, 100]:
    plt.plot(x, chi2.pdf(x, n - 1), label = "df = "+str(n - 1))
    #reaches a peak at n - 2
    plt.legend(loc='upper right')
plt.title('$\chi^2\ \mathrm{Distribution}$')
plt.show()
```
