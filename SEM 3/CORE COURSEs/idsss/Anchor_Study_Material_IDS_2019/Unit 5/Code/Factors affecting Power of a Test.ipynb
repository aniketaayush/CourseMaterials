```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from scipy.stats import norm

n = 30
sigma = 10
alpha = 0.05

# H0 : mu >= 50
# H1 : mu < 50 // left tailed test; Power : area to left

mu0 = 50
```

```python
#moving in the direction of the alternative : power increases
z_nd = norm.ppf(0.05)
se = sigma/sqrt(n)
critical_xbar = z_nd * se + mu0
power = []
for i in range(40, mu0):
    z_ad = (critical_xbar - i)/se
    power.append(norm.cdf(z_ad))
plt.plot(range(40, mu0), power)
plt.show()
```

```python
#changing sample size for various alternatives : increase in sample size : increases power
n_list = [10,30, 100, 500]

z_nd = norm.ppf(0.05) #aplha = 0.05 ; left tailed test => critical z_score = -1.645

for sample_size in n_list:
    se = sigma/sqrt(sample_size) #standard error
    critical_xbar = z_nd * se + mu0
    #calculating power for different alternatives
    power = []
    for i in range(40, mu0):
        z_ad = (critical_xbar - i)/se
        power.append(norm.cdf(z_ad))
    label_value = "n = "+str(sample_size)
    plt.plot(range(40, mu0), power, label=label_value)
    plt.legend(loc = "lower left")
plt.show()
```

```python
#changing level of the test : increase in alpha : increases power

alpha_list = [0.01, 0.05, 0.10]

se = sigma/sqrt(n)

for level in alpha_list:
    z_nd = norm.ppf(level)
    critical_xbar = z_nd * se + mu0
    power = []
    for i in range(40, mu0):
        z_ad = (critical_xbar - i)/se
        power.append(norm.cdf(z_ad))
    label_value = "alpha = "+str(level)
    plt.plot(range(40, mu0), power, label=label_value)
    plt.legend(loc = "lower left")
plt.show()
```

```python
#changing standard deviation : decrease in sd : increases power
sigma_list = [5, 10, 20]

z_nd = norm.ppf(alpha)

for sd in sigma_list:
    se = sd/sqrt(n)
    critical_xbar = z_nd * se + mu0
    power = []
    for i in range(40, mu0):
        z_ad = (critical_xbar - i)/se
        power.append(norm.cdf(z_ad))
    label_value = "sigma = "+str(sd)
    plt.plot(range(40, mu0), power, label=label_value)
    plt.legend(loc = "lower left")
plt.show()
```
