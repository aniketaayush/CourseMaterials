# 1. Cosine Distance 
You will be using this a lot in the course and ahead if you pursue data science. Cosine distance of two vectors is given by dot product of two vectors divided by product of their magnitudes. 

```python
import math

def magnitude(a):
	squared_sum = 0
	for i in a:
		squared_sum += i*i
	return math.sqrt(squared_sum)

def cosine_similarity(a, b):
	ans = 0
	for i in range(len(a)):
		ans += (a[i]*b[i])
	return ans/(magnitude(a)*magnitude(b))


a =  (5,7,8,7)
b = (5,7,7,8)
print(cosine_similarity(a,b))
```

# 2. Jaccard Distance

Jaccard distance between two sets is given by ratio of cardinality of intersection to cardinality fo union

```python
from math import*
 
def jaccard_similarity(x,y):

    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
 
print(jaccard_similarity([0,1,2,5,6],[0,2,3,5,7,9]))
```

# 3. K-Nearest Neihbours

Knn is a simple supervised classification algorithm. In the following example we will see how the classification depends on the value of k

```python
# importing libraries

import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, neighbors
from mlxtend.plotting import plot_decision_regions
```

```python
# this function plots the knn results. 
def knn_comparison(data, k):
    x = data[['X','Y']].values
    y = data['class'].astype(int).values
    clf = neighbors.KNeighborsClassifier(n_neighbors=k)
    clf.fit(x, y)
    # Plotting decision region
    plot_decision_regions(x, y, clf=clf, legend=2)
    # Adding axes annotations
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Knn with K='+ str(k))
    plt.show()
```

```python
# plot for various k values
data1 = pd.read_csv('data_worksheet_2/ushape.csv')
for i in [1,5,20,30,40,80]:
    knn_comparison(data1, i)
```

```python
data2 = pd.read_csv('data_worksheet_2/linearsep.csv')
for i in [1,5,20,30,40,60]:
    knn_comparison(data2, i)
```
